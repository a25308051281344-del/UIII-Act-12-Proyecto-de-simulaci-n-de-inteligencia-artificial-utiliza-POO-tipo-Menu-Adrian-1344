```python
print("========================================================")
print("            IA CHEF MICHELIN")
print("========================================================")


class Chef:

    def __init__(self, estrellas, especialidad, restaurante,
                 brigada, menu, anos):

        self.estrellas = estrellas
        self.especialidad = especialidad
        self.restaurante = restaurante
        self.brigada = brigada
        self.menu = menu
        self.anos = anos


lista_chefs = []


# FUNCION AGREGAR
def agregar_chef():

    print("\n=========== AGREGAR CHEF ===========")

    estrellas = input("Estrellas Michelin: ")
    especialidad = input("Especialidad culinaria: ")
    restaurante = input("Restaurante: ")
    brigada = input("Brigada a cargo: ")
    menu = input("Menú insignia: ")
    anos = input("Años en cocina: ")

    chef = Chef(estrellas, especialidad,
                restaurante, brigada,
                menu, anos)

    lista_chefs.append(chef)

    print("\nChef agregado correctamente.")


# FUNCION LISTAR
def listar_chefs():

    print("\n=================== LISTA DE CHEFS ===================")

    print("==============================================================================================================")
    print("| No | ESTRELLAS | ESPECIALIDAD | RESTAURANTE | BRIGADA | MENÚ INSIGNIA | AÑOS EN COCINA |")
    print("==============================================================================================================")

    contador = 1

    for chef in lista_chefs:

        print("|",
              str(contador).ljust(2), "|",
              chef.estrellas.ljust(10), "|",
              chef.especialidad.ljust(13), "|",
              chef.restaurante.ljust(12), "|",
              chef.brigada.ljust(8), "|",
              chef.menu.ljust(15), "|",
              chef.anos.ljust(16), "|")

        contador += 1

    print("==============================================================================================================")


# FUNCION ACTUALIZAR
def actualizar_chef():

    listar_chefs()

    posicion = int(input("\nNúmero del chef a actualizar: "))

    if posicion <= len(lista_chefs):

        print("\n======= NUEVOS DATOS =======")

        lista_chefs[posicion - 1].estrellas = input("Nuevas estrellas Michelin: ")
        lista_chefs[posicion - 1].especialidad = input("Nueva especialidad: ")
        lista_chefs[posicion - 1].restaurante = input("Nuevo restaurante: ")
        lista_chefs[posicion - 1].brigada = input("Nueva brigada: ")
        lista_chefs[posicion - 1].menu = input("Nuevo menú insignia: ")
        lista_chefs[posicion - 1].anos = input("Nuevos años en cocina: ")

        print("\nChef actualizado correctamente.")

    else:
        print("\nNúmero incorrecto.")


# FUNCION BORRAR
def borrar_chef():

    listar_chefs()

    posicion = int(input("\nNúmero del chef a borrar: "))

    if posicion <= len(lista_chefs):

        lista_chefs.pop(posicion - 1)

        print("\nChef eliminado correctamente.")

    else:
        print("\nNúmero incorrecto.")


# MENU PRINCIPAL
while True:

    print("\n================ MENÚ PRINCIPAL ================")
    print("1. Agregar chef")
    print("2. Listar chefs")
    print("3. Actualizar chef")
    print("4. Borrar chef")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCION 1
    if opcion == "1":
        agregar_chef()

    # OPCION 2
    elif opcion == "2":
        listar_chefs()

    # OPCION 3
    elif opcion == "3":
        actualizar_chef()

    # OPCION 4
    elif opcion == "4":
        borrar_chef()

    # OPCION 5
    elif opcion == "5":

        print("\nGracias por usar IA Chef Michelin.")
        print("Programa finalizado.")
        break

    # OPCION INVALIDA
    else:
        print("\nOpción no válida.")
```
